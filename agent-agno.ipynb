## Завдання 1. Агент з інструментами веб-пошуку

```python
# Імпортуємо головний клас для створення інтелектуальних агентів
from agno.agent import Agent
# Імпортуємо модуль для інтеграції з локальним сервером Ollama
from agno.models.ollama import Ollama
# Імпортуємо інструмент безкоштовного та анонімного пошуку в мережі Інтернет
from agno.tools.duckduckgo import DuckDuckGoTools

# Ініціалізуємо агента-дослідника
web_agent = Agent(
    name="Локальний Дослідник",
    # Задаємо локальну модель, яку попередньо завантажили через 'ollama run llama3.2'
    model=Ollama(id="llama3.2"),
    # Передаємо список інструментів, до яких агент матиме безпосередній доступ
    tools=[DuckDuckGoTools()],
    # Формулюємо правила поведінки та обмеження для ШІ
    instructions=[
        "Відповідай виключно українською мовою.",
        "Обов'язково вказуй посилання на джерела інформації (URLs), які повернув інструмент пошуку.",
    ],
    # Дозволяємо використання красивого маркдаун-форматування (списки, жирний текст)
    markdown=True,
    # Вмикаємо дебаг, щоб у терміналі бачити лог генерації JSON для Tool Calls
    debug_mode=True
)

# Запускаємо тест для першого запиту на перевірку динамічних новин
print("--- Запит 1: Актуальні новини ---")
web_agent.print_response(
    "Які головні події відбулися у сфері технологій за останній тиждень?",
    stream=True,
    stream_intermediate_steps=True
)
```

## Завдання 2. Створення власного інструменту

```python
from agno.agent import Agent
from agno.models.ollama import Ollama
# Імпортуємо декоратор @tool, який перетворює звичайну Python-функцію на інструмент для ШІ
from agno.tools import tool

# Обов'язково пишемо docstring (опис), бо за ним модель розуміє, навіщо ця функція потрібна
@tool
def get_current_datetime(timezone: str = "Europe/Kyiv") -> str:
    """Отримує поточну дату та час для заданого часового поясу.

    Args:
        timezone: Назва часового поясу (наприклад, 'Europe/Kyiv', 'America/New_York').

    Returns:
        Рядок з поточною датою та часом у форматі YYYY-MM-DD.
    """
    from datetime import datetime
    import pytz # Бібліотека для роботи з часовими поясами

    try:
        tz = pytz.timezone(timezone)
        now = datetime.now(tz)
        return f"Поточний час у зоні {timezone}: {now.strftime('%Y-%m-%d %H:%M:%S')}"
    except Exception as e:
        return f"Помилка визначення часового поясу: {str(e)}"

@tool
def generate_secure_password(length: int = 12) -> str:
    """Генерує випадковий безпечний пароль, що містить літери, цифри та спецсимволи.

    Args:
        length: Довжина пароля (за замовчуванням 12).

    Returns:
        Згенерований безпечний пароль для захисту акаунтів.
    """
    import random
    import string

    if length < 6: # Захист від занадто коротких паролів
        length = 6
    # Об'єднуємо всі можливі символи: літери, цифри та знаки пунктуації
    characters = string.ascii_letters + string.digits + string.punctuation
    password = "".join(random.choice(characters) for _ in range(length))
    return f"Згенеровано надійний пароль довжиною {length} символів: {password}"

# Створюємо агента і реєструємо наші дві Python-функції як доступні інструменти
custom_agent = Agent(
    name="Агент Спеціальних Інструментів",
    model=Ollama(id="llama3.2"),
    tools=[get_current_datetime, generate_secure_password],
    instructions=[
        "Ти повинен обов'язково НАЙПЕРШЕ ВИКЛИКАТИ інструменти у форматі JSON, якщо користувач просить дізнатися час або створити пароль.",
        "Не вигадуй відповіді самостійно, використовуй лише те, що повертають функції.",
        "Відповідай українською мовою."
    ],
    markdown=True,
    debug_mode=True
)

# Надсилаємо комплексний запит, який змусить модель виконати обидва інструменти одночасно
custom_agent.print_response(
    "Підкажи, котра зараз година в Лондоні (Europe/London)? Також згенеруй мені надійний пароль на 16 символів.",
    stream=True,
    stream_intermediate_steps=True
)
```

## Завдання 3. Створення команди агентів (Team)

```python
from agno.agent import Agent
# Імпортуємо клас Team для координації та оркестрації мультиагентних систем
from agno.team.team import Team
from agno.models.ollama import Ollama
from agno.tools.duckduckgo import DuckDuckGoTools

# Створюємо першого вузькоспеціалізованого агента — пошуковця
market_researcher = Agent(
    name="Market_Researcher",
    role="Глибокий збір та агрегація ринкових даних з Інтернету",
    model=Ollama(id="llama3.2"),
    tools=[DuckDuckGoTools()],
    instructions=[
        "Шукай фактичні ринкові дані, тренди та статистику.",
        "Формулюй пошукові запити (query) ВИКЛЮЧНО чистою англійською або чистою українською мовою.",
        "НІКОЛИ не змішуй англійські та українські літери в одному слові (наприклад, жодних 'тendenції').",
        "Передавай виключно знайдені факти без власних висновків українською мовою."
    ],
    markdown=True
)

# Створюємо другого агента — текстового аналітика
business_analyst = Agent(
    name="Business_Analyst",
    role="Професійний аналіз сирих даних та структурування бізнес-звітів",
    model=Ollama(id="llama3.2"),
    instructions=[
        "Ретельно аналізуй сирі факти, які тобі передав Market_Researcher.",
        "Заборонено вигадувати дані, якщо їх немає у звіті дослідника.",
        "Формуй звіт за структурою: 1. Актуальний стан, 2. Ризики та можливості, 3. Фінальний висновок.",
        "Пиши чіткою діловою українською мовою."
    ],
    markdown=True
)

# Об'єднуємо їх під керуванням лідера команди (Team)
business_team = Team(
    name="Локальна Консалтингова Команда",
    # Перелічуємо учасників команди
    members=[market_researcher, business_analyst],
    model=Ollama(id="llama3.2"),
    delegate_to_all_members=False, # Вимикаємо хаотичний спам усім підряд
    # Прописуємо алгоритм взаємодії (оркестрацію) для координатора
    instructions=[
        "1. Скоординуй роботу: спочатку дай команду Market_Researcher зібрати дані за запитом користувача.",
        "2. Перевір, щоб запит до дослідника був сформульований чітко (наприклад: 'Electric cars market trends in Ukraine').",
        "3. Отриманий масив текстів від дослідника передай агенту Business_Analyst для написання фінального звіту."
    ],
    # Дозволяємо агентам бачити історію повідомлень та репліки один одного
    share_member_interactions=True,
    # Відображаємо у терміналі проміжні текстові відповіді кожного з учасників
    show_members_responses=True,
    markdown=True,
    debug_mode=True
)

# Викликаємо команду для виконання складного аналітичного бізнес-завдання
business_team.print_response(
    "Зроби аналіз ринку електромобілів в Україні: які зараз тренди та перспективи розвитку?",
    stream=True
)
```

## Завдання 4. Агент з можливостями Reasoning

```python
from agno.agent import Agent
from agno.models.ollama import Ollama

# Створюємо аналітичного агента для вирішення логічних задач
reasoning_agent = Agent(
    name="Локальний Логічний Аналітик",
    model=Ollama(id="llama3.2"),
    # Замість стандартних інструментів, ми конфігуруємо модель на покрокове мислення за допомогою промпт-інжинірингу
    instructions=[
        "Ти — експерт із системного мислення та теорії графів. Розв'язуй задачі строго покроково.",
        "Перед тим як дати фінальну відповідь, обов'язково заповни блок внутрішніх міркувань за структурою:",
        "  - КРОК 1: Аналіз поточного стану, обмежень та небезпечних комбінацій.",
        "  - КРОК 2: Побудова дерева рішень (варіанти ходів туди і назад).",
        "  - КРОК 3: Перевірка безпеки кожного етапу та вибір єдиного вірного шляху.",
        "Спілкуйся українською мовою. Будь максимально критичним до своїх кроків."
    ],
    markdown=True,
    debug_mode=True
)

# Задаємо класичну математичну/логічну задачу про переправу, яка вимагає побудови чіткої послідовності дій
reasoning_agent.print_response(
    """
    У фермера є вовк, коза і капуста. Йому потрібно переправити їх через річку в човні, 
    але в човен крім фермера вміщується лише один об'єкт (або тільки вовк, або тільки коза, або тільки капуста).
    Якщо залишити вовка з козою без нагляду, вовк з'їсть козу. Якщо залишити козу з капустою, коза з'їсть капусту.
    Розпиши покроковий алгоритм безпечної переправи з детальним обґрунтуванням кожного кроку.
    """,
    stream=True,
    stream_intermediate_steps=True
)
```
